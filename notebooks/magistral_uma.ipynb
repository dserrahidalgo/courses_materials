## 3. Los trucos

Los trucos que sabe hacer Uma son, en cierto modo, funciones que ha ido consolidando Uma a base de práctica. Una función puede tomar dos tipos principales de argumentos: los **posicionales** y los **keyword**.

Los argumentos posicionales son obligatorios, es decir, cuando llamemos a la función, debemos incluirlos dentro de los paréntesis. 


### Sienta
Uma sabe muy bien lo que hacer cuando le dicen "sienta". Basta con que le digan su nombre y la acción.

En la función sienta, **'nombre'** es un argumento posicional, por lo que siempre tiene que ser nombrado cuando se llama a la funcion, de lo contrario dará error.

Es buena costumbre incluir lo que se llama una docstring en la función, que básicamente es un comentario (normalmente con triples dobles comillas) donde indicamos en cristiano lo que hace dicha función:

```python
def sienta (nombre):
    """Esta función nos devuelve la imagen de Uma sentada.
    Argumentos:
    -nombre: el nombre de nuestra compañera canina. 
    """
    print(nombre, ', sienta!!')
    return imagen('./fotos_uma/sienta.png')
```

```python
sienta('Uma')
```

```python
# Si no mencionamos el argumento posicional, vamos a tener un error.
sienta()
```

### Tumba. Argumentos keywords
Y también cuando le dicen tumba. De hecho, se tiene tan bien aprendido el truco que no necesita que la llamen por su nombre. Ha aprendido que cuando alguien próximo a ella le dice "tumba", se refiere a ella.

En este caso, el argumento 'nombre' es un argumento keyword, o argumentos palabra clave. Estos argumentos se definen dando un valor por defecto al argumento, de forma que si al llamar a la función no lo incluímos, se utilizará el valor por defecto. Por lo tanto, nombrar esos argumentos es voluntario.


```python
def tumba (nombre='Uma'):
    print(nombre, ', tumba!!')
    imagen('./fotos_uma/tumba.png')
    pass;
```

```python
tumba()
```

Uma es una perra muy sociable y cariñosa. Sea como fuere, ella ha aprendido a dar la "patita". No porque crea que es una forma natural de saludar, sino porque normalmente dar la patita suele ir acompañado de un premio. Y eso mola.

```python
def patita (nombre, la_otra='no'):
    if nombre == 'Uma':
        print (nombre + ', patita!!')
        imagen('./fotos_uma/patita.png')
        
        if la_otra=='si':
            print('la otra!!')
            imagen('./fotos_uma/la_otra.png')
    else:
        print('¿me estás hablando a mí?')
        return imagen('./fotos_uma/paso_de_ti.png')
```

### Choca. Número arbitrario de argumentos
En el ejemplo anterior hemos creado una función bastante flexible. No obstante, estamos limitados a siempre llamarla con 1 argumento posicional y otro keyword. Esto es una clara limitación... ¿Y si quiero meter un número arbitrario de elementos en nuestra función?

Para ello existe lo que se llama *args. Es básicamente una sentencia que permite:

-Que una función tome tantos argumentos posicionales como se quiera.

-Iterar por este argumento args, el cual es un iterable que contiene todos los argumentos posicionales, para generar la lógica que queramos en el cuerpo de nuestra función.

### 3.4. Sube
La mejor amiga de Uma es Gata. Gata, como su nombre indica, es una gata. De tanto vivir juntas, a Uma se le han pegado algunas cosas de Gata, como el gusto de subir a sitios poco habituales. Como es evidente, Uma no es Gata, ni tampoco es una gata, sino una perra, por lo que su capacidad llegar a un sitio depende de varios factores.

```python
def ladra(*args, fuegos_artificiales='no'):
    if fuegos_artificiales =='no':
        print('Uma está tranquila')
        imagen('./fotos_uma/tumbada.png')
        
    if fuegos_artificiales =='si':
        imagen('./fotos_uma/fuegos_artificiales.png', 5, 5)
        for ladrido in args:
            print(ladrido * 2)
        
        imagen('./fotos_uma/ladra.png', 5, 5)
```

```python
ladra('guau! ','guauu! ','guauuuu!! ','guauuuuu! ','GUAUUUUUUUU!!!! ', fuegos_artificiales='no')
```

```python
ladra('guau! ','guau! ','guauuu!! ','guauuuuu! ','GUAUUUUUUUU!!!! ', fuegos_artificiales='si')
```

## Trucos. Número arbitrario de keywords argumentos
el parámetro especial ** kwargs en una función se usa para pasar, de forma opcional, un número variable de argumentos keywords.

El parámetro recibe los argumentos como un diccionario.

Al tratarse de un diccionario, el orden de los parámetros no importa. Los parámetros se asocian en función de las claves del diccionario.

```python
def trucos(**kwargs):
    for truco, foto in kwargs.items():
        print(truco.upper())
        imagen('./fotos_uma/'+ foto)
```

```python
trucos(patita='patita.png', choca = 'choca.png', sube ='sube.png',lado='lado.png')
```

```python
def sube (altura, cansada = False):
    "altura en centímetros"
    print('Uma, salta!!')
    
    if (altura  > 50 and altura <= 150):

        if (altura >= 100 and cansada==False):
            imagen('sienta.png')
        elif (altura >= 100 and cansada==True):
            imagen('tumba.png')
        elif ((altura < 100) and (cansada == False or cansada == True)):
            imagen('sienta.png')
    
    elif altura > 150:
        raise Exception('Imposible')
    
    else:
        raise Exception ('No hay anda para saltar')
        
```
