### 3.3. Patita
Uma es una perra muy sociable y cariñosa. Si los saludos entre humanos le parece una cosa rara (y más ahora con esto del covid19), más raro le parece aún los saludos entre perros y humanos. ¿A qué viene tanta insistencia en pedir la "patita"? ¿Qué tendrá esa palabra que les haga tanta ilusión?
Sea como fuere, ella ha aprendido a dar la "patita". No porque crea que es una forma natural de saludar, sino porque normalmente dar la patita suele ir acompañado de un premio. Y eso mola.

### 3.4. Sube
La mejor amiga de Uma es Gata. Gata, como su nombre indica, es una gata. De tanto vivir juntas, a Uma se le han pegado algunas cosas de Gata, como el gusto de subir a sitios poco habituales. Como es evidente, Uma no es Gata, ni tampoco es una gata, sino una perra, por lo que su capacidad llegar a un sitio depende de varios factores.

```python
def sube (altura, cansada = False):
    "altura en centímetros"
    print('Uma, salta!!')
    
    if (altura  > 50 and altura <= 150):

        if (altura >= 100 and cansada==False):
            imagen('sienta.png')
        elif (altura >= 100 and cansada==True):
            imagen('tumba.png')
        elif ((altura < 100) and (cansada == False or cansada == True)):
            imagen('sienta.png')
    
    elif altura > 150:
        raise Exception('Imposible')
    
    else:
        raise Exception ('No hay anda para saltar')
        
```

```python
sube(180, True)
```

```python
import random
 
participantes = ['javier', 'manuel', 'pedro', 'sara', 'cristina',
                 'belen', 'marcos', 'fernando', 'juan', 'ismael','luisma']
 

```

```python
def parejas(nombres):
    while len(nombres) > 0:
        
        pareja = random.sample(nombres, 2)
        print(pareja)
        
        for nombre in list(nombres):
            if nombre in pareja:
                nombres.remove(nombre)
        
        if len(nombres) ==3:
            pareja = random.sample(nombres, 3)
            print(pareja)
            break
            
                
           
```

```python
parejas(participantes)
```
